```python
import time
import json
print("#########################################")
print("***********1.Add product*****************")
print("***********2.Remove product**************")
print("***********3.Show items******************")
print("***********4.Shopping********************")
print("Note: Plz visit show items before going for\nshowing to make sure that product is avilable")
user_choice=int(input("Enter your choice: "))
if(user_choice == 1):
  id=input(print("$Enter the product id: " ))
  name=input(print("$Enter the product name: "))
  quat=int(input(print("$Enter the quantity: ")))
  price=int(input(print("$Enter the Price: "))) 
  mfd=input(print("$Enter the Manufacturing date: ")) 
  exp=input(print("$Enter the Expiry_date: "))
  newproduct={id:{'product_name':name , 'quantity':quat , 'Price':price ,'Manufacturing_date':mfd , 'Expiry_date':exp}}
  fd=open("record.json",'r')
  txt = fd.read()
  fd.close()
  records=json.loads(txt)
  records.update(newproduct)
  js=json.dumps(records)
  fd=open("record.json",'w')
  fd.write(js)
  fd.close()
  print("The product updated successfully")
elif (user_choice == 2):
  delid=input(print("$Enter product id to be deleted"))
  fd=open("record.json",'r')
  txt = fd.read()
  fd.close()
  records=json.loads(txt)
  del records[delid]
  js=json.dumps(records)
  fd=open("record.json",'w')
  fd.write(js)
  fd.close()
  print("product removed successfully")
elif (user_choice == 3):
  fd=open("record.json",'r')
  txt = fd.read()
  fd.close()
  records=json.loads(txt)
  print("&&&&&& List of products &&&&&&&")
  for i in records.keys():
       print("&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&")
       print("Product id: ",i)
       print("Product name: ",records[i]['product_name'])
       print("Quantity avilable: ",records[i]['quantity'])
       print("Price: ",records[i]['Price'])
elif (user_choice == 4):
  coname=input("Enter your name: ")
  conum=input("Enter your Ph.no: ")
  proid=input("Enter product id: ")
  qua=int(input("Enter product quantity: "))
  fd=open("customer.json",'r')
  text = fd.read()
  fd.close()
  customers=json.loads(text)
  fd=open("record.json",'r')
  txt = fd.read()
  fd.close()
  records=json.loads(txt)
  points=(records[proid]['Price']*qua)/10
  for i in customers.keys():
      if (conum == i ):
        customers[i]['points']= int(customers[i]['points']+points)
        b=0
      else:
        newcustomer={conum:{'name':coname,'points':int(points)}}
        b=1
  if(b==1):
    {
     customers.update(newcustomer)
    }

  hh=json.dumps(customers)
  ha=open("customer.json",'w')
  ha.write(hh)
  ha.close()
  fd=open("customer.json",'r')
  text = fd.read()
  fd.close()
  customerss=json.loads(text)
  for i in records.keys():
      if(proid == i):
          records[i]['quantity']=records[i]['quantity']-qua 
          print("\n\n\n")
          print("############################################")
          print("############## HK Company ##################")
          print("############# Welcomes you #################")
          print("Your shoping details: ")
          print("producr ID:         ",proid)
          print("Product Name:       ",records[i]['product_name'])
          print("Product Price:      ",records[i]['Price'])
          print("Product Quantity:   ",qua)
          print("Manufacturing date: ",records[i]['Manufacturing_date'])
          print("Expiry date:        ",records[i]['Expiry_date'])
          print("------------------------------------------")
          print("Billing amount:     ",records[proid]['Price']*qua)
          print("------------------------------------------")
          print("Your wallet point:  ",customerss[conum]['points'])
          print("Billing time:       ",time.ctime())
          print("#############################################")
          print("********HAPPY SHOPPING ",coname,"********")
          fd=open("sale.json",'r')
          mxt = fd.read()
          fd.close()
          sale=json.loads(mxt)
          newsale={str(len(sale)+1):{'product_id':proid,'product_name':records[i]['product_name'],'quantity':qua,'Amount':records[proid]['Price']*qua}}
          sale.update(newsale)
          js=json.dumps(sale)
          fd=open("sale.json",'w')
          fd.write(js)
          fd.close()
          ms=json.dumps(records)
          ww=open("record.json",'w')
          ww.write(ms)
          ww.close()
```
